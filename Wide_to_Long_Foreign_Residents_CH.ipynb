```python
# import libraries
import pandas as pd
import numpy as np
```

```python
fresidents_df = pd.read_excel("Foreign_permanent_residents_CH.xlsx")
```

```python
fresidents_df.head(10)
```

```python
fresidents_df.info()
```

```python
# convert from wide to long with pdmelt, first create list for the years
year_list = list(fresidents_df.loc[:,fresidents_df.columns != "Citizenship"])
fresidents_long = pd.melt(fresidents_df, id_vars = "Citizenship", value_vars = year_list, var_name = "Year", value_name = "Population")
```
